```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the dataset
file_path = r"C:\Users\User\Downloads\World_Bank\API_SP.POP.TOTL_DS2_en_csv_v2_900.csv"
data = pd.read_csv(file_path, skiprows=4)  # Skipping metadata rows
```

```python
# Step 2: Inspect the dataset
print("First few rows of the dataset:")
print(data.head())
print("\nAvailable columns:")
print(data.columns)
```

```python
# Step 3: Filter the data for visualization
# Select the year for visualization
year = "2021"  
if year in data.columns:
    population_data = data[["Country Name", year]].dropna()

    # Sort by population and select the top 10 countries
    top_countries = population_data.sort_values(by=year, ascending=False).head(10)
    # Step 4: Create a bar chart
    plt.figure(figsize=(12, 6))
    plt.bar(top_countries["Country Name"], top_countries[year], color="skyblue")
    plt.title(f"Top 10 Populous Countries in {year}")
    plt.xlabel("Country")
    plt.ylabel("Population")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig("bar_chart.png")  # Save the bar chart as an image
    plt.show()

    # Step 5: Create a histogram for population distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(population_data[year], bins=20, color="purple", kde=True)
    plt.title(f"Population Distribution in {year}")
    plt.xlabel("Population")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.savefig("population_histogram.png")  # Save the histogram as an image
    plt.show()
else:
    print(f"Year {year} not found in the dataset. Please check the columns.")
```
